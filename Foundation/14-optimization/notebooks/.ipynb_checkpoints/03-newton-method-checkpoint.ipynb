# Introduction to Newton's Method

In numerical analysis, Newton's method (also known as the Newtonâ€“Raphson method), named after Isaac Newton and Joseph Raphson, is a method for finding successively better approximations to the roots (or zeroes) of a real-valued function. It is one example of a root-finding algorithm.

The following mathematical algorithm describe how the root is generated by the iterative process in the formula. We here omit the mathematical proof in getting showing the formula can achieve the result

# Python Implementation of newton's method

```python
def f(x):
    return 9 - x*(x-10)

def f_dev(x):
    return -2*x +10

guess = -10

for val in range(1,10):
    nextguess = guess - f(guess)/f_dev(guess)
    print(nextguess)
    guess = nextguess
```

```python
# using scipy.optimize
from scipy import optimize

def f(x):
    return 9 - x*(x-10)

root = optimize.newton(f, -10)

print(root)
```
