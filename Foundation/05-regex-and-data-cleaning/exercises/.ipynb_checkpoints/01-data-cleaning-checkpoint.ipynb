# Open Rice

** 1.) ** We scrapped a data set from open rice but it needs cleaning:

* Remove duplicate data
* Use a regex to clean the review col
* Remove the trailing commar from the adress col


```python
import pandas as pd
import re
```

```python
openrice = pd.read_csv("data/open-rice.csv")
```

```python
df = openrice.drop_duplicates(subset=['address','name'])
print(f"Shape before drop dups: {openrice.shape}, after: {df.shape}")
```

```python
# Cleaning the review column
s = "(123 Reviews)"
def get_number(x):
    m = re.search(r'\d+',x)
    return m.group()

#get_number(s)
df['reviews_count'] = df.number_of_reviews.apply(get_number)
df.head(5)
```

```python
# %load solutions/openrice_1.py
```

### JHO's Note
when re.search() found nothing, we will return None instead of a Match object 

```python
# let's fix the address too
s = 'Shop J-K., 200 Hollywood Road'
def trim_coma(x):
    m = re.search(r'(.+?),$', x)
    str_ans = ''
    if m is None:
        str_ans = x
    else:
        str_ans = m.group(1)
    return str_ans

#trim_coma(s)
df['clean_address'] = df.address.apply(trim_coma)
df.head(5)
```

```python
df['clean_address'] = openrice.address.apply(
                        lambda x: re.sub(',$','',x)
                        )
df.head(5)
```

There's to many resturants types, this can hinder the analysis. Lets make a new column for HK Style Resturants.

```python
def hkstyle_to_numeric(x):
    if x == 'Hong Kong Style':
        return 1
    else:
        return 0
```

```python
# we can use .apply to generate new columns for it
df['HKrestuarant'] = df.food_type.apply( lambda x: 'HKresturant' if hkstyle_to_numeric(x) == 1 else 'Non-HKresturant')
df.head(3)
```

** 3.) ** *Bonus* - In collumns that should be numerical use regular expressions to get the numbers and convert the columns from strings to numerical.

```python
import re 
```

```python
# %load solutions/ricacorp_3.py
```

```python
ricacorp.dtypes
```

```python
df.loc[:,'pay per month'].sample(10)
```

```python
s = '1st Pay︰$13.25M'
m = re.search(r'\$(\d+\.?\d*)',s)
m.group(1)
```

```python
cleaning_dict= {
    '1st payment': r'\$(\d+\.?\d*)',
    'loan %': r'(\d+\%)',
    'pay per month': r'\$(\d+\.?\d*)',
    'area': r'(\d?\,*\d+)Ft'
}
```

```python
for ikey in cleaning_dict:
    df[ikey] = ricacorp[ikey].apply(
        lambda s: re.search(cleaning_dict[ikey] ,s).group(1)
        )
df.head(3)
```
