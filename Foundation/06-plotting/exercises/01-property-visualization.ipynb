# Property Visualization

```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
import numpy as np
sns.set(color_codes=True) #overide maplot libs ugly colours.
```

```python
df = pd.read_csv('../data/Hong_Kong_Clean.csv')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df.date.apply( lambda x: x.year )
df.head(3)
```

**Plot a graph showing the price per square feet over the years.**

```python
fig, ax = plt.subplots(figsize= (13,8))
sns.barplot(
           data = df,
           x = 'year',
           y = 'sqft_price'
)
```

```python
fig, ax = plt.subplots(figsize=(13,8))
```

**Plot a graph showing the area per square feet against each district.**

```python
fig, ax = plt.subplots(figsize= (8,20))
sns.barplot(
           data = df,
           x = 'sqft_area',
           y = 'district')
```

```python
plt.figure(figsize=(8,20))
```

**Plot a graph showing relationship between floor and the price**

```python
fig, ax = plt.subplots(figsize= (13,8))
sns.barplot(
           data = df,
           x = 'floor',
           y = 'sqft_price',
            palette = 'Blues_d')
```

```python
plt.figure(figsize=(13,8))
```
