# Matplotlib

Matplotlib is for many plotting libraries in python. It's very powerfull and can create all sort of plots, givining the prorammer complete control over the apperance of the plots. Such control can sometimes however be a burder, since you have to specify may things to create what may be like a simple plot.

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
#magic needed to make plots show in matplotlib
%matplotlib inline
```

```python
mpl.rcParams['figure.figsize'] =(13,8) #set default figure size
```

# Figures and Axes

A figure objects represent the whole image (or firgure) whereas and the axis represents a individual plot. A figure object may contain a single axis object or multiple.

```python
fig = plt.figure()  #nothing shows since its empty figure
```

```python
fig = plt.figure(figsize=(16,8)) #specify size in cm, width and height
ax = fig.add_subplot(111) #1 row, 1 col, 1 ?
```

We modify the axis object to add labels, title ,etc. 

```python
fig = plt.figure(figsize=(13,8))
ax = fig.add_subplot(111) 
ax.set(xlim=[0.5, 4.5],
       ylim=[-2, 8], 
       title='An Example Axes',
       ylabel='Y-Axis',
       xlabel='X-Axis')
plt.show()
```

The above code could also have been written `set_something` instead, this can get slightly tedious but it gives us more control, for example bellow I've bumped up the size of the title.

```python
fig = plt.figure(figsize=(13,8))
ax = fig.add_subplot(111) 
ax.set_xlim([0.5, 4.5])
ax.set_ylim([-2, 8])
ax.set_title('An Example Axes',size=25)
ax.set_ylabel('Y-Axis')
ax.set_xlabel('X-Axis')
```

Since pretty much anytime we create a figure we'll also want to make a axis object, an alternative  syntax is avaliable which will create both.

```python
fig,ax = plt.subplots()
```

We can create multiple axis in a single figure easily aswell

```python
fig,axes = plt.subplots(1,2) #rows, columns
```

```python
axes # axes is and array of axis objects
```

# Plotting

Up untill now we've only made axis object, lets draw something on them.


```python
# create some dummy data
a = np.linspace(1,10,10)
b = a**2
```

`.scatter` takes two arguremnts a x and and a y argument and draws a point.

```python
fig,ax = plt.subplots()
ax.scatter(a,b) 
```

We can also draw lines using `.plot` . If we only pass `plot` a single argument then the index will be used for the x-axis and the list's values for the y-axis.

```python
fig,axes = plt.subplots(1,2,figsize=(13,8))
axes[0].plot(a) #take as s single
axes[1].plot(1/a ,b) # or double argument
```

## Axes methods vs. pyplot

Interestingly, just about all methods of an *Axes* object exist as a function in the *pyplot* module (and vice-versa). For example, when calling `plt.xlim(1, 10)`, *pyplot* calls `ax.set_xlim(1, 10)` on whichever *Axes* is "current". Here is an equivalent version of the above example using just pyplot.

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

When we just have a single plot, almost all of the axes object functions also exist on the pyplot module, forexample 

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='x')
plt.xlim(0.5, 4.5)
plt.show()
```

```python
import numpy as np
```

```python
xs, ys = np.mgrid[:4, 9:0:-1]
markers = [".", "+", ",", "x", "o", "D", "d", "", "8", "s", "p", "*", "|", "_", "h", "H", 0, 4, "<", "3",
           1, 5, ">", "4", 2, 6, "^", "2", 3, 7, "v", "1", "None", None, " ", ""]
descripts = ["point", "plus", "pixel", "cross", "circle", "diamond", "thin diamond", "",
             "octagon", "square", "pentagon", "star", "vertical bar", "horizontal bar", "hexagon 1", "hexagon 2",
             "tick left", "caret left", "triangle left", "tri left", "tick right", "caret right", "triangle right", "tri right",
             "tick up", "caret up", "triangle up", "tri up", "tick down", "caret down", "triangle down", "tri down",
             "Nothing", "Nothing", "Nothing", "Nothing"]
fig, ax = plt.subplots(1, 1, figsize=(7.5, 4))
for x, y, m, d in zip(xs.T.flat, ys.T.flat, markers, descripts):
    ax.scatter(x, y, marker=m, s=100)
    ax.text(x + 0.1, y - 0.1, d, size=14)
ax.set_axis_off()
plt.show()
```

## Linestyles
Line styles are about as commonly used as colors. There are a few predefined linestyles available to use. 

linestyle          | description
-------------------|------------------------------
'-'                | solid
'--'               | dashed
'-.'               | dashdot
':'                | dotted


```python
t = np.arange(0.0, 5.0, 0.2)
plt.figure(figsize=(13,8))
plt.plot(t, t, '-', t, t**2, '--', t, t**3, '-.', t, -t, ':');
```

It is a bit confusing, but the line styles mentioned above are only valid for lines. Whenever you are dealing with the linestyles of the edges of "Patch" objects, you will need to use words instead of the symbols. So "solid" instead of "-", and "dashdot" instead of "-.". This issue will be fixed for the v2.1 release and allow these specifications to be used interchangably.

```python
fig, ax = plt.subplots(1, 1)
ax.bar([1, 2, 3, 4], [10, 20, 15, 13], ls='solid', ec='r', lw=5)
plt.show()
```

# Colours 

Matplot lib has some usefull colour maps build it, these can be accessed form `plt.cm`.

```python
from seaborn import load_dataset
```

```python
df= load_dataset('iris')
X = df.iloc[:,[1,2]].values

#convert strings to numbers
y = df.iloc[:,-1]
d = { label : i  for i, label in enumerate(y.unique())}
y = y.map(d).values
```

With a scatter plot we can use the `c` parameter to pass the label colour for  each point.

```python
plt.scatter(X[:,0],X[:,1], c=y)
```

```python
from cycler  import cycler
```

```python
cm = plt.cm.rainbow
n_colours = 5
color_cycler = cycler('color',cm(np.arange(n_colours)/n_colours))
```

```python
x = np.arange(0.0, 5.0, 0.2)
fig,ax = plt.subplots(figsize=(13,8))
ax.set_prop_cycle(color_cycler)
for i in range(1,5):
    plt.plot(x,x**i);
```
