# Sklearn


The purpose of this notebook is to get you familair with the sklearn API for clustering.



** 1. ** Apply each Hierarachical, Kmeans and DBSCAN to the bellow dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import make_moons, make_circles
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
X1, y1 = make_blobs(n_samples=1000, centers=5, n_features=2,cluster_std = 0.7 , random_state=0)
X2, y2 = make_moons(n_samples =1000,noise=0.1)
X3, y3 = make_circles(1000,noise=0.01)
```

```python
fig, axes = plt.subplots(1,3,figsize=(24,8))

for i,X in enumerate([X1,X2,X3]):
    axes[i].scatter(X[:,0],X[:,1])
```

```python
models = [KMeans, AgglomerativeClustering, DBSCAN]
n_clusters = [5,2,2]
data = [X1,X2,X3]
```

```python
results = { 'KMeans': [] ,'HC':[], 'DBSCAN':[]}
```

```python
for k,model in zip(['KMeans','HC','DBSCAN'],models):
    for n, X in zip(n_clusters, data):
        results[k].append(model(n).fit_predict(X))
        
```

** 2. ** In each case evalulate how well each clustering methods did by plotting?

```python
fig, axes = plt.subplots(3,3,figsize=(10,10))
for i,model in enumerate(['KMeans','HC','DBSCAN']):
    for j,X in enumerate(data):
        colours = results[model][j]
        axes[i][j].set_title(model)
        axes[i][j].scatter(X[:,0],X[:,1], c=colours)
        
```

```python
cg = sns.clustermap(X1)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, distance
from scipy.cluster.hierarchy import fcluster
```

```python
plt.figure(figsize=(20,10))
dendrogram(cg.dendrogram_row.linkage,no_labels=True)
max_d =  2.5
plt.axhline(y=max_d, c='k')
```

```python
labels = fcluster(cg.dendrogram_row.linkage,max_d,criterion='distance')
```
