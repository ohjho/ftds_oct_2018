**Q1: This is a more challenging question. You have done BeautifulSoup in the last projects. Now implement a function directly where you pass a URL and an empty list to get all the links on that page.**

```python
#@title
import urllib2
from bs4 import BeautifulSoup

def getAllDoxyDonkeyPosts(url,links):
    request = urllib2.Request(url)
    response = urllib2.urlopen(request)
    soup = BeautifulSoup(response)
    for a in soup.findAll('a'):
        try:
            url = a['href']
            title = a['title']
            if title == "Older Posts":
                print title, url
                links.append(url)
                getAllDoxyDonkeyPosts(url,links)
        except:
            title = ""
    return

blogUrl = "http://doxydonkey.blogspot.in"
links = []
getAllDoxyDonkeyPosts(blogUrl,links)
```

**Q2: Let's move to the next big question. Implement a function that takes a URL and returns the posts in it.**

```python
#@title
def getDoxyDonkeyText(testUrl):
    request = urllib2.Request(testUrl)
    response = urllib2.urlopen(request)
    soup = BeautifulSoup(response)
    mydivs = soup.findAll("div", {"class":'post-body'})
    
    posts =[]
    for div in mydivs:
        posts+=map(lambda p:p.text.encode('ascii', errors='replace').replace("?"," "), div.findAll("li"))
    return posts
```

**Q3: Now print the posts for every URL using a for loop and the function you just made**

```python
doxyDonkeyPosts = []
for link in links:
    doxyDonkeyPosts+=getDoxyDonkeyText(link)
    
doxyDonkeyPosts
```

**Q5: Make a vectorizer variable with max_df as 0.5 and min_df as 0.2 and stop_words as english. **

```python
#@title
vectorizer = TfidfVectorizer(max_df=0.5,min_df=2,stop_words='english')
```

**Q9: Now let's import KMeans and create an object with 3 clusters. **

```python
#@title
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 100, n_init = 1, verbose = True)
```

**Q10: Now fit the data into this**

```python
#@title
km.fit(X)
```
