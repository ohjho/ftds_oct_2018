**Auto Summarizing text**

In this unit we will auto-summarize text by getting it through beautiful soup and then tokenize it with NLP. 


Start with importing Beautiful Soup and urllib2. 

```python
#@title
import urllib2
from bs4 import BeautifulSoup
```

Pick a random article URL. 

```python
#@title
articleURL = "https://www.washingtonpost.com/news/the-switch/wp/2016/10/18/the-pentagons-massive-new-telescope-is-designed-to-track-space-junk-and-watch-out-for-killer-asteroids/"
```

**Q1: Open the link using Beautiful soup. Use "html.parser" as the argument.**

```python
#@title
page = urllib2.urlopen(articleURL).read().decode('utf8','ignore') 
soup = BeautifulSoup(page,"html.parser")
soup
```

**Q2: Do a soup.find() on the article variable.**

```python
#@title
soup.find('article')
```

```python
#@title
soup.find('article').text
```

**Q4: Join the article based on it's spaces.**

```python
#@title
text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
text
```

**Q5: Encode the text using ASCII and replace the unwarranted characters.**

```python
#@title
text.encode('ascii', errors='replace').replace("?"," ")
```

**Q6: Form a function for whatever you did above so one can pass a URL and get an encoded ASCII text in return.**

```python
#@title
def getTextWaPo(url):
    page = urllib2.urlopen(url).read().decode('utf8')
    soup = BeautifulSoup(page,"html.parser")
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    return text.encode('ascii', errors='replace').replace("?"," ")
```

**Q7: Run the above function to test it**

```python
#@title
text = getTextWaPo(articleURL)
```

**Q8: Import stopwords, punctuation, sent_tokenize and word_tokenize**

```python
#@title
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

**Q9: Tokenize the sentences**

```python
#@title
sents = sent_tokenize(text)
sents
```

**Q10: Tokenize the words and make them lower case**

```python
#@title
word_sent = word_tokenize(text.lower())
word_sent
```

**Q11: Show the stopwords.**

```python
#@title
_stopwords = set(stopwords.words('english') + list(punctuation))
_stopwords
```

**Q12: Print the tokenized words if they are not in stopwords.**

```python
#@title
word_sent=[word for word in word_sent if word not in _stopwords]
word_sent
```

**Q13: Print the frequency distribution**

```python
#@title
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

**Q14: Print the n largest words by frequency**

```python
from heapq import nlargest
nlargest(10, freq, key=freq.get)
```

**Q15: Using default dict, rank the tokenized words as per the frequency**

```python
#@title
from collections import defaultdict
ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
            
ranking
```

**Q16: Print the 4 largest sentence indexes**

```python
#@title
sents_idx = nlargest(4, ranking, key=ranking.get)
sents_idx
```

**Q17: Print the sentences**

```python
#@title
[sents[j] for j in sorted(sents_idx)]
```

**Q18: For whatever we just did above, make a function summarize so we input a text and n and get the top frequency sentences.**

```python
#@title
def summarize(text, n):
    sents = sent_tokenize(text)
    
    assert n <= len(sents)
    word_sent = word_tokenize(text.lower())
    _stopwords = set(stopwords.words('english') + list(punctuation))
    
    word_sent=[word for word in word_sent if word not in _stopwords]
    freq = FreqDist(word_sent)
    
    
    ranking = defaultdict(int)
    
    for i,sent in enumerate(sents):
        for w in word_tokenize(sent.lower()):
            if w in freq:
                ranking[i] += freq[w]
             
        
    sents_idx = nlargest(n, ranking, key=ranking.get)
    return [sents[j] for j in sorted(sents_idx)]
```

```python
summarize(text,3)
```
