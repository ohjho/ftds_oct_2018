### 1. Import Data
csv here is from ML studio

```python
import pandas as pd

data = pd.read_csv( 'data/automobile.csv')   # this data set is from Azure ML studio
print(data.dtypes)
data.head(3)
```

### 2. Clean Missing Data and Feature Engineering

```python
l_col_to_use = ['engine-size','horsepower','peak-rpm','city-mpg', 'highway-mpg','price']
df = data.dropna(subset = l_col_to_use).loc[:, l_col_to_use]
df.head(3)
```

### 3. Split Data
into training and test data  
[another great post](https://towardsdatascience.com/train-test-split-and-cross-validation-in-python-80b61beca4b6) on test splitting data

```python
from sklearn.model_selection import train_test_split

l_predictors = l_col_to_use[:-1]
X_train, X_test, y_train, y_test = train_test_split(df.loc[:,l_predictors], df.price, test_size = 0.3, random_state = 0)

print(f'Test data shapes: {X_train.shape} {y_train.shape}')
print(f'Test data shapes: {X_test.shape} {y_test.shape}')
```

### 4. Train the Model
A decision tree [tutorial](https://stackabuse.com/decision-trees-in-python-with-scikit-learn/)

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
model = lm.fit(X_train, y_train)

# let's compare it with decision tree
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
tmodel = regressor.fit(X_train, y_train)
```

### 5. Score the Model

```python
%matplotlib inline
from matplotlib import pyplot as plt
prediction = lm.predict(X_test)
score = lm.score(X_test, y_test)

print(f'Linear Regression model Score: {score}')
plt.plot(y_test, prediction, '.')
plt.xlabel('True Values')
plt.ylabel('Predictions')
```

```python
tprediction = regressor.predict( X_test)
tscore = regressor.score( X_test, y_test)

print(f'Decision Tree Regression model Score: {tscore}')
plt.plot(y_test, tprediction, '.')
plt.xlabel('True Values')
plt.ylabel('Predictions')
```

### 6. Deploy the model

```python
x_input = {'engine-size': [92], 'horsepower': [62],'peak-rpm': [4800],'city-mpg': [27],'highway-mpg': [32]}
x_input = pd.DataFrame.from_dict(x_input)
test = lm.predict(x_input)
test
```

## Let's try a Classification Experiment

### 3. Split Data
now we will predict the make with the six column above

```python
l_col_to_use = ['engine-size','horsepower','peak-rpm','city-mpg', 'highway-mpg','price', 'make']
df = data.dropna(subset = l_col_to_use[:-1]).loc[:, l_col_to_use]
df.head(3)
```

```python
l_predictors = l_col_to_use[:-1]
X_train, X_test, y_train, y_test = train_test_split(df.loc[:,l_predictors], df.make, test_size = 0.3, random_state = 0)

print(f'Test data shapes: {X_train.shape} {y_train.shape}')
print(f'Test data shapes: {X_test.shape} {y_test.shape}')
```

### 4. Train the model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()

cmodel = classifier.fit( X_train, y_train)
```

### 5. Score the model

```python
cpredict = classifier.predict(X_test)
cscore = classifier.score(X_test, y_test)

pdata = { 'cpredict': cpredict, 'y_test': y_test}
out_df = pd.DataFrame.from_dict(pdata)
out_df['correct'] = out_df.cpredict == out_df.y_test

print(f'Decision Tree Classifier model Score: {cscore}')
print(f'{"{:.2%}".format(sum(out_df.correct)/len(out_df))} correct')
out_df
```
