```python
import pandas as pd
```

```python
url_path = 'https://raw.githubusercontent.com/raybellwaves/cfanalytics/master/Data/Affiliate_list.csv'
affiliate_list = pd.read_csv(url_path)
```

```python
affiliate_list.head()
```

```python
df = affiliate_list.drop(columns=['Unnamed: 0']).set_index('Affiliate_id')
df.head(3)
```

```python
grouped = df.groupby('Country')['Affiliate_name'].agg({'Affiliate_count':'count'})
grouped.sort_values(by = 'Affiliate_count', ascending = False)
```

```python
df_us = df[df.Country=='United States']
g_df_us = df_us.groupby('State')['Affiliate_name'].agg({'Affiliate_count':'count'})
g_df_us.sort_values(by = 'Affiliate_count', ascending =False)
```

```python
df_ca = df[df.Country=='Canada']
g_df_ca = df_ca.groupby('State')['Affiliate_name'].agg({'Affiliate_count':'count'})
g_df_ca.sort_values(by = 'Affiliate_count', ascending =False)
```

```python
df_ch = df[df.Country=='C']
g_df_ch = df_ca.groupby('State')['Affiliate_name'].agg({'Affiliate_count':'count'})
g_df_ch.sort_values(by = 'Affiliate_count', ascending =False)
```

```python
df_ch = df[(df.Country=='China') | (df.Country == 'Hong Kong') | (df.Country == 'Macau')]
g_df_ch = df_ch.groupby('State')['Affiliate_name'].agg({'Affiliate_count':'count'})
g_df_ch.sort_values(by = 'Affiliate_count', ascending =False)
```

### Let's Plot the Affiliates Locations on a Map using Folium

The map tiles matter! Read [this](https://python-graph-gallery.com/288-map-background-with-folium/)

And this [Kaggle Tutorial](https://www.kaggle.com/daveianhickey/how-to-folium-for-maps-heatmaps-time-analysis) is also very useful

```python
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
l_latlon = df[['Latitude','Longitude']].dropna().values[:]
start_loc = list(l_latlon[0])
```

```python
# init map
map_aff = folium.Map( location = start_loc, zoom_start= 2, tiles = 'Mapbox Control Room')

# for i in l_latlon:
#     folium.Marker( list(i), popup='London Bridge').add_to(map_aff)

# heat map
heat_data = [list(i) for i in l_latlon]
HeatMap(heat_data).add_to(map_aff)

# show map
map_aff
```
