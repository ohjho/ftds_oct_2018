# Beginner Machine Learning
let's predict stock returns using regression model

## 1. Import Data
we use Panda's data reader

```python
import pandas as pd
from pandas_datareader import data as web
from datetime import datetime as dt

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

lsymbols = [
    'SPY',
    'VXX',
    'USO',
    'GLD',
    'QQQ',
    'IWM',
    'SLV'
]

sdate = dt(2017,1,1)
yhoo_data = web.DataReader( lsymbols, 'yahoo', sdate, dt.now())
yhoo_data.head(3)
```

## 2. Clean Data and Feature Engineering
We need to convert prices into Stock Returns

### First Massage the Data a bit

```python
data = yhoo_data.stack()
data.reset_index(inplace = True)
data.head(3)
```

### Then transform the data
1. Create df of returns for each stock
2. recombine the stock return dfs into one big df

```python
from functools import reduce

l_col_to_keep = ['Date', 'Symbols', 'Adj Close', 'Volume']
r_dfs = []

# 1
for sym in lsymbols:
    r_df = data[ data.Symbols == sym ].loc[:, l_col_to_keep]
    r_df = r_df.set_index('Date')
    r_df[f'r({sym})'] = r_df['Adj Close']/ r_df['Adj Close'].shift(1) - 1
    # of just do df.pct_change()
    r_dfs.append( r_df.iloc[1:,:])

# 2
df = reduce( lambda x, y : pd.concat([x,y], axis =1),
            [ r_df.iloc[:,-1] for r_df in r_dfs ]
           )

df.head(3)
```

## 3. Exploratory Data Analysis

```python
sns.pairplot(df)
```

```python
m_corr = df.corr()
plt.figure( figsize = (16,8))
sns.set_context('paper', font_scale = 1.2)
sns.heatmap( m_corr, annot= True)
```
